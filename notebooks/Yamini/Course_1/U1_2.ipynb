## <a id = 'var'>Variables</a>

<b>Variables</b> are containers or memory locations for storing data values.<br>
In the example below, the variable, name, stores a string that represents your name and the variable, number, holds your favorite integer. <br>
The print statement can be enhanced to include literal and variable information. Different parts of the print statement are separated. by commas. Variable names are not enlosed in double quotes.<br>

```python
name ="Type your name here"
number = 8
print("Hi I am ", name, "and ", number, "is my favorite number")
```

The variables are created the moment you assign a value to them.<br>
Variable names have to adhere to the following rules:
- Can be of any length
- Can be upper case or lower case letters or a mixture of both
- Can contain digits, but not as the first character
- Can contain the underscore charater

## <a id='dt'>Data Types </a>

A variable is assigned a <b>data type</b> based on the data it holds.<br>
Run the following cell to see the type of data each of the variables hold.<br>

```python
name ="Type your name here"
number_int = 8
number_float=9.0
true_bool = True
print(type(name))
print(type(number_int))
print(type(number_float))
print(type(true_bool))
```

Python supports different data types - both <u>single data values</u> and <u>collections of data</u>. <br>
We will focus on basic data types that hold a single value in this lesson.<br>
The ones we use most often to store a single data value are:<br><br>
<b>Numbers<b>
- <b>Integers:</b> -3, 0, 4 etc<br>
- <b>Floating points:</b> -3.15, +2.75 etc<br>

<b>Text</b>
The text data type needs to be enclosed in double quotes. <br>
- <b>String: </b>"Hello"<br>

<b>Boolean</b> 
- This data type only holds a true or false value and is used to make logical decisions in the programs.<br>
- <b>bool: </b>True or False<br>

```python
int1 = -3
float1 = 3.0
string1 = "Hello"
bool1 = True

print("Integer: ", int1, "Type: ", type(int1))
print("Float: ", float1, "Type: ", type(float1))
print("String: ", string1, "Type: ", type(string1))
print("Boolean: ", bool1, "Type: ", type(bool1))
```

More complex data types like lists, tuples, arrays, sets and dictionaries hold a collection of these simple data types.<br><br>
<b> Variable Names</b><br>
It is good programing practice to use variables names that describe their use. This makes it easier for others to understand the code you have written.<br><br>
<b>Changing data type of variables </b><br>
The data type of a variable can be changed by changing its value as illustrated by the code below.<br>

```python
gen = 3
print(type(gen))
gen = 3.0
print(type(gen))
gen ="String"
print(type(gen))
```

## <a id='op'>Operators</a>
We discuss some of the basic operators supported by Python in the following sections. 

### <a id='aop'>Arithmetic operators</a>

Arithmetic operations follow the order of operation (PEMDAS) rules of math.<br><br>
The ones that require a little extra explanation are the '%' and '//'.<br><br>
'%' gives the modulus or remainder after division. Example 5%3=2. <br>
A simple application of this would be to check if a number is odd or even by inspecting the remainder after diving the number by 2. <br><br>
'//' gives the integer floor part of the division by rounding down. Example: 7//4 =1.<br><br>
The '+' can also be used to concatenate (add one string to the end of the other) strings.<br>

### <a id ='rop'>Relational Operators</a>
In order to write any meaningful computer applications, we need the power to make decisions based on data and relational or comparison operators allow us to do this.<br>
They return a value of 'True' or 'False' based on the operation performed.<br>


```python
a=2
b=3
c=3
print(a<b)
print(a==b)
print(a!=b)
print(b==c)
print(b>=c)
print(b<c)
```

### <a id='lop'>Logical Operators</a>


```python
print(2<3 and 4==4)
print(2>3 and 4==4)
print(2<3 or 4!=4)
print(2>3 or 4==4)
print(not(2<3 and 4==4))
```
