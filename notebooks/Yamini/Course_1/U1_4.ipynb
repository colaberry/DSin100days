## <a id='for'>For Loops</a>
One of the key advantages of a computer is the fact that it can perform repetitive tasks at great speeds. The <b>for loop</b> is one such structure that allows us to harness this power. It repeatedly executes a certain block multiple times based on start and stop conditions.<br>
As with the <b>If</b> statment, any lines of code inside these blocks of code need to be indented.<br>

```python
# Simple for loop that prints the integers 1 though 4
for i in range(1,5):
    print(i)
```

As you can see from the above, <b>range</b> never includes the stop number in its result.<br>
The range takes parameters range(start, stop, [step]). The step is an optional parameter.<br>
The default start is 0 and need not be specified.<br>
The default step is +1 and need not be specified either.<br>
The range can also be used to step through the elements of a list.<br>
The membership operator <b>'in'</b> is used here.<br>


```python
# Simple for loop that prints the integers 0 though 4 without specifying start
for i in range(5):
    print(i)
```

```python
# Simple loop to print only even numbers from 0 through 8
for i in range(0,9,2):
    print(i)
```

```python
# Accessing all elements in a list
list_soda = ["pepsi", "sprite", "mountain dew", "pepsi", "coke"]
for s in list_soda:
    print(s)
```

Another way of obtaining the same results would be to use a counter and accessing the elements by index position .<br>

```python
list_soda = ["pepsi", "sprite", "mountain dew", "pepsi", "coke"]
for i in range(0, len(list_soda)):
    print("i= ", i , "element =", list_soda[i])
```

Accessing by index position gives us the flexibility of skipping certain elements by using different start, stop or step parameters.<br>
Always remember that the index positions start at 0 and end at (n-1), where n is the number of elements in the list.<br>

```python
list_soda = ["pepsi", "sprite", "mountain dew", "pepsi", "coke"]
# Print only 3rd to 5th elements 
print("Print only 3rd to 5th elements ")
for i in range(2, 4):
    print("i= ", i , "element =", list_soda[i])
```

You can even count backwards in the <b>for</b> loop. In this case, the start value will be greater than the stop value.

```python
list_soda = ["pepsi", "sprite", "mountain dew", "pepsi", "coke"]
for i in range(len(list_soda),0,-1):
    print("i= ", i , "element =", list_soda[i-1])
```

## <a id ='freq'>Frequency</a>
20 students in a class were asked what their favorite soda was, and the information recorded is shown below. <br>
["pepsi", "sprite", "mountain dew", "pepsi", "coke", "coke", "fanta", "pepsi", "coke", "coke", "sprite", "canada dry", "coke", "pepsi", "coke", "sprite", "mountain dew", "canada dry", "fanta", "coke"]<br>
The teacher asks them to find how many students like each soda. <br>

```python
# Data collected from students
list_soda = ["pepsi", "sprite", "mountain dew", "pepsi", "coke", "coke", "mountain dew", "pepsi", "coke", "coke", "sprite", "canada dry", "coke", "pepsi", "coke", "sprite", "mountain dew", "canada dry", "fanta", "coke"]
# Set the counter for each type of drink to 0
pepsi=0 # pepsi count
coke=0 # coke count
md=0 # mountain dew count
cd=0# canada dry count
fanta=0 # fanta count

# The for loop goes through each element of the loop and checks it against the names.
# If a match is found, the count of that drink is incremented (increased) by 1.
for s in list_soda:
    if(s=="pepsi"):
        pepsi=pepsi+1
    if(s=="coke"):
        coke=coke+1
    if(s=="mountain dew"):
        md = md+1
    if(s =="canada dry"):
        cd=cd+1
    if(s == "fanta"):
        fanta = fanta+1
print(pepsi, " students like pepsi")
print(coke, " students like coke")
print(md, " students like mountain dew")
print(cd, " students like canada dry")
print(fanta, " students like fanta")
```

Looking at the output above it is clear that the order of preference is coke, pepsi, mountain, dew, canada dry and fanta.<br>

## <a id='sets'> Sets </a>
<b>Sets</b> are a collection of elements which are <b>unique</b> i.e. do not contain duplicates.<br>
We can use this property of sets and built in count functions to shorten our code to find the frequency of certain events.<br>
Using the set(list_name) we can obtain a unique list of elements<br>

```python
# Data collected from students
list_soda = ["pepsi", "sprite", "mountain dew", "pepsi", "coke", "coke", "mountain dew", "pepsi", "coke", "coke", "sprite", "canada dry", "coke", "pepsi", "coke", "sprite", "mountain dew", "canada dry", "fanta", "coke"]
# Make a list of the unique soda names. We use the set function to return unique elements
list_un = set(list_soda)
print("The unique list of sodas contain", len(list_un), "and has elements:" ,list_un)
# Access each unique soda 
for s in list_un:
    # count(name of soda) automatically counts all instances of a particular soda for us
    print (s, ":",list_soda.count(s) )
```
