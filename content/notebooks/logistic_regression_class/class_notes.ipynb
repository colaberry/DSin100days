```python
# import statements 
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np 
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


from sklearn import model_selection


column_names = np.array(load_iris().feature_names+["target"])
data_columns = np.c_[load_iris().data, load_iris().target]

iris_data = pd.DataFrame(data=data_columns, columns=column_names)
```

```python

iris_subset = iris_data[iris_data["target"] <=1.0]
X = iris_subset["sepal length (cm)"].values
y = iris_subset["target"].values

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
%matplotlib notebook 

plt.scatter(x_train, y_train)
plt.show()
```

```python
column_names = np.array(load_iris().feature_names+["target"])
data_columns = np.c_[load_iris().data, load_iris().target]

iris_data= pd.DataFrame(data=data_columns, columns=column_names)

#single column
iris_first_two = iris_data.iloc[np.where(iris_data["target"] <=1.0)]

X = iris_first_two[["sepal length (cm)"]].values
y = iris_first_two["target"].values

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)

log_reg = LogisticRegression(solver="lbfgs" , random_state=3)

# fitting the model
log_reg.fit(x_train,y_train)

# prediction on the test dataset 
pred_y = log_reg.predict(x_test)

# calculating the accuracy score 
accuracy_score(pred_y, y_test)
```

```python
%matplotlib notebook 

model_line_x = np.linspace(np.min(x_train),np.max(x_train), 100)


exponent_value  = ((log_reg.coef_[0][0]*model_line_x )+ log_reg.intercept_[0])

denominator = 1 + np.exp(-exponent_value)


logistic_function =  1.0 / denominator
plt.figure()
plt.scatter(x_test, y_test)
plt.plot(model_line_x, logistic_function, color="red")
plt.show()
```

```python
column_names = np.array(load_iris().feature_names+["target"])
data_columns = np.c_[load_iris().data, load_iris().target]

iris_data= pd.DataFrame(data=data_columns, columns=column_names)

#single column
iris_first_two = iris_data.iloc[np.where(iris_data["target"] <=1.0)]

X = iris_first_two.iloc[:,0:4].values
y = iris_first_two["target"].values

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)

log_reg = LogisticRegression(solver="lbfgs" , random_state=3)

# fitting the model
log_reg.fit(x_train,y_train)

# prediction on the test dataset 
pred_y = log_reg.predict(x_test)

# calculating the accuracy score 
accuracy_score(pred_y, y_test)
```

  # Multinomial logistic regression 

```python
X = iris_data.iloc[:,0:4].values
y = iris_data["target"].values

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
log_reg = LogisticRegression(solver="lbfgs" , random_state=3, multi_class="multinomial", max_iter=1000)
log_reg.fit(x_train, y_train)
pred_y = log_reg.predict(x_test)
accuracy_score(pred_y, y_test)
```
