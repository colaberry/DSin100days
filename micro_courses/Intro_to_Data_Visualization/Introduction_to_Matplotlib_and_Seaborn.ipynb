# Introduction to Matplotlib & Seaborn

## Data Visualization

Data Visualization is the presentation of data in a pictorial or graphical format. an effective way to communicate your findings about the dataset. The goals of visualization are:

* To gain powerful insights into data that aren't obvious with statistics. 
* Business and non-technical audience are able to comprehend the results.
* Useful to build stories so peer data scientists can understand the analysis. 

In this micro course, we'll introduce two important data visualization libraries "matplotlib" and "seaborn".

## 1. Matplotlib

Matplotlib is a basic graphing library in python. It is the most commonly used package by data scientists for visualization. Many of the more advanced visualization tools use matplotlib as the base library. Hence you will need to develop some basic familiarity with how to use matplotlib. 

Let us learn to use matplotlib by plotting the sine curve. The sine of a variable can be computed using the numpy library with np.sin() function.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 100)
plt.plot(x, np.sin(x))
plt.xlabel('Theta')
plt.ylabel('y=sin(x)')
plt.show()
```

### 1.1 Scatter Plot

A scatter plot shows the relationship between two numerical variables. Each member of the dataset gets plotted as a point whose (x, y) coordinates relates to its values for the two variables. 

Let's walk through a very simple example using two numpy arrays. Note in this example we added a magic line "**%matplotlib inline**", which will give us static plots. We can also use "**%matplotlib notebook**", which can provide interactivity of the plot. (The downside of "%matplotlib notebook" is that all the subsequent plots will be plotted in this interactive plot and overlie on each other. To avoid this, we should plot individual figures. We can do this by adding plt.figure() before each plot.)

````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# %matplotlib notebook

x = np.linspace(0, 5, 11)
y = x ** 2

plt.scatter(x, y)
plt.xlabel('X values')
plt.ylabel('X values')
plt.title('Scatter Plot')
plt.show()
````

### 1.2 Distribution Plot

 
 Plotting histograms is a common task in many data science problems. Histograms show the frequency count for a range of values in case of a discrete variable. For a continuous variable, the data has to be binned to generate a histogram.  
```python 
plt.hist(x,bins=30)
```
 can be used to plot histograms in python. 
 
 The argument 'bins' can be used to control the number of bins. The more the number of bins, the smaller the size of the bins. 'x' is a 1 dimensional numpy array of data. 
 
 Let's see the the histogram for a continuous distribution of values 

````python
x = np.random.normal(size=100)
plt.hist(x, bins=30)
plt.show()
````

### 2. Seaborn

Seaborn is a Python visualization library based on matplotlib. It provides a high-level interface for drawing attractive statistical graphics. The graphs produced are also aesthetic and convey more information useful for Data Scientists.

Seaborn comes with built-in data sets! Let's see how we can import tips dataset and analyze it using seaborn.

````python
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
````

### 2.1 _Joint Plot_

Seaborn's joint plot also shows the relationship between two numerical variables. It is equivalent to matplotlib's scatterplot.

Let's see how people 'tip' w.r.t the 'total_bill' in the tips dataset.

````python
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
````

### 2.2 Seaborn: Distribution Plot

Seaborn has a number of plots to visualize distributions. The most important one is the dist plot which shows distribution of a univariate set of observations. Other distribution plots are rugplot and kdeplot.

Let's see the distribution of 'total_bill in the tips dataset.

````python
sns.distplot(tips['total_bill'], kde=False)
plt.show()
````

### 2.3 Pair Plot

Seaborn's pairplot is more like a scatter plot, but it is used to show multiple relationships between different pairings of the variables in a dataset. It shows the relationship between all numeric values in a dataset, all non-numeric values are automatically ignored.

Let's create a pairplot to visualize the variability of all numeric fields in the 'iris' dataset.

````python
iris = sns.load_dataset('iris')
print(iris.head())
sns.pairplot(iris)
````

### 2.4 KDE Plot

KDE plots are Kernel Density Estimation Plots. It is a lot like a histogram but gives the probability distribution. This means that instead of plotting actual values it estimates the probability density of a continuous variable. KDEs are especially useful in providing a continuous representation of histograms. 

Let's create a KDE plot to see the probability distribution of 'tip' in tips dataset.

````python
sns.kdeplot(tips['tip'])
plt.xlabel('Tip Amount')
plt.ylabel('Probability Density')
plt.title('Probability density plot of Tip Amount')
````

### 2.5 regplot and lmplot

Seaborn has many built-in capabilities for regression plots. These regression plots are primarily intended to add a visual guide that helps to emphasize patterns in a dataset during exploratory data analyses. 

Let's see what a regplot() can do to our tips dataset.

````python
sns.regplot(x="total_bill", y="tip", data=tips)
````

The solid line in the plot is the linear regression model fit. The translucent band lines, however, describe a bootstrap confidence interval generated for the estimate.

An lmplot() is used in the same way as regplot() and it does the same fitting. The main difference between them is that regplot() is an axes-level function, so it draws directly onto an axes, while lmplot() is a figure-level function and creates its own figure. This means that regplot() can coexist in a figure with other kinds of plots while lmplot() needs to occupy an entire figure.

### 2.6 boxplot() and  violinplot()

A __boxplot()__ is used to show the distribution of data within categories.

For example, it can be used to visualize the distribution of 'total_bill' for each 'day' and for each 'time' within each 'day'.

````python
sns.boxplot(x="day", y="total_bill", hue="time", data=tips)
````

A better visualization is the __viloinplot()__. It gives the categories and also the Kernel Density Curve of other numerical variable within each category.

````python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips)
````
