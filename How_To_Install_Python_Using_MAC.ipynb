

# How to Download and Install Anaconda/Jupyter Notebook Using MacOS


## Step1


First, you have to check the systme requirements here according to your Operating System: <br>
https://conda.io/docs/user-guide/install/index.html <br>

Scroll down to the **"System Requirements"** and make sure you meet the requirements. Don't worry if you do not have Python installed already. Installing Anaconda automatically installs Python.
Then go to **"Regular Installation"** section and choose your Operating System (Windows or macOS or Linux). This will redirect you to another webpage. <br>


Since I will show you how to install Anaconda using MacOS, please follow this link. <br>
https://www.anaconda.com/download/#macos <br>
